```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/raw/spotify_tracks.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
genres = [
    "doom-metal",
    "thrash-metal",
    "death-metal",
    "symphonic-metal"
]

filtered = df[df["track_genre"].isin(genres)]

filtered["track_genre"].value_counts()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=filtered,
    x="track_genre"
)

plt.xticks(rotation=45)

plt.show()
```
